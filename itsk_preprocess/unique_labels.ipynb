```python
import os 
import numpy as np
import SimpleITK as sitk
```

```python
baseDir = r"/home/konata/Dataset/TED_MRI/T2/mask/seg_cut/"
files=os.listdir(baseDir)
unique=dict()
uniques=list()
for file in files:
    nunique=0
    uniques=[]
    img=sitk.ReadImage(os.path.join(baseDir,file))
    img=sitk.GetArrayFromImage(img)
    for i in range(img.shape[0]):
        if len(np.unique(img[i,:,:]))>nunique:
            nunique=len(np.unique(img[i,:,:]))  #找到同一冠状面上最大标签数
    for i in range(img.shape[0]):
        if len(np.unique(img[i,:,:]))==nunique: #找到满足最大标签数的冠状面索引
            uniques.append(i+1) #i+1方便对照itsk软件查看（itsk从1开始），python切片索引为i（从0开始）
            print(i+1)
    print(file,nunique)
    unique[file]=[nunique,uniques]
    #key：文件名
    #value：[冠状面最大标签数,[最大标签数冠状面索引+1,…]]
```

```python
import json
with open('/home/konata/Dataset/TED_MRI/T2/mask/seg_unique.json', 'w') as file:
    json.dump(unique, file) #存入json
unique
```
