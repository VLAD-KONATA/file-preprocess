```python
#随机获取50-50TAO/正常数据

df=pd.read_csv('/home/konata/Dataset/TAO_CT/testmask.csv',header=None,names=['id','label'])
df.drop([0],axis=0,index=None,columns=None,inplace=True)
sample_0=df[df['label']=='0'].sample(50,random_state=42)
sample_1=df[df['label']=='1'].sample(50,random_state=42)

samples=pd.concat([sample_0,sample_1])
samples=samples.reset_index(drop=True)

output_folder='/home/konata/Dataset/TAO_CT/50-50cut/'
output_folder2='/home/konata/Dataset/TAO_CT/50-50mask/'
for file_id in samples['id']:
    shutil.copy(os.path.join('/home/konata/Dataset/TAO_CT/testcut',file_id),os.path.join(output_folder,file_id))
    shutil.copy(os.path.join('/home/konata/Dataset/TAO_CT/testmask',file_id),os.path.join(output_folder2,file_id))
    
samples.to_csv('/home/konata/Dataset/TAO_CT/50-50cut.csv',index=False)
```

```python
dir='/home/konata/Dataset/TAO_CT/test/'
list=os.listdir(dir)
#list=['3546828_2020-01-03.nii.gz']
a=set()
for file in list :
    try:
        img=nib.load(dir+file)
        affine=img.affine
        data=img.get_fdata()
        orentation=aff2axcodes(affine)
        if orentation==('L', 'A', 'S'):
            affine[:,1]*=-1
            flipped=np.flip(data,axis=1)
            new_img=nib.Nifti1Image(flipped,affine)
            nib.save(new_img,dir+file)
            print('las '+file+' processed')
        elif orentation==('L', 'S', 'P'):
            affine[:,[1,2]]=affine[:,[2,1]]
            swap=np.swapaxes(data,1,2)
            new_img=nib.Nifti1Image(swap,affine)
            nib.save(new_img,dir+file)
            print('lsp '+file+' processed')
        elif orentation==('L','S','A'):
            affine[:,1]*=-1
            flipped=np.flip(data,axis=1)
            new_img=nib.Nifti1Image(flipped,affine)
            nib.save(new_img,dir+file)
            print('lsa '+file+' processed')
        elif orentation==('L','I','A'):
            affine[:,1]*=-1
            affine[:,2]*=-1
            flipped=np.flip(data,axis=(1,2))
            new_img=nib.Nifti1Image(flipped,affine)
            nib.save(new_img,dir+file)
            print('lia '+file+' processed')
    except:
        print(file)
        shutil.move(os.path.join(dir,file),os.path.join('/home/konata/Dataset/TAO_CT/ERROR',file))
```
