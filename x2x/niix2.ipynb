```python
import os
import numpy as np
import SimpleITK as sitk
from pathlib import Path
```

```python
def niix2(inpath,outpath,name):
    img=sitk.ReadImage(os.path.join(inpath,name))
    img=sitk.Image(img)
    array=sitk.GetArrayFromImage(img)
    shape=array.shape
    zero=np.zeros((shape[0]*2-1,shape[1],shape[2]),dtype=np.float32)
    for i in range(shape[0]):
        zero[i*2,:,:]=array[i,:,:]
    spacing=(0.3125,0.3125,3.85/2)
    outs=sitk.GetImageFromArray(zero)
    outs.SetDirection(img.GetDirection())
    outs.SetOrigin(img.GetOrigin())
    outs.SetSpacing(spacing)
    sitk.WriteImage(outs,os.path.join(outpath,name))
    print(f"Saved NifTI file to {os.path.join(outpath,name)}")
```

```python
inpath='/home/konata/Dataset/TED_MRI/TOM500/label/'
outpath='/home/konata/Dataset/TED_MRI/TOM500/labelx2/'
Path(outpath).mkdir(exist_ok=True, parents=True)
files=os.listdir(inpath)
for file in files:
    niix2(inpath,outpath,file)
```
