```python
import os
import SimpleITK as sitk
import numpy as np
import json
import tqdm
import pandas as pd
```

```python
#读取(unique<=9)
baseDir = r'/home/konata/Dataset/TED_MRI/T2/seg_slice_un2d_all_Jun_17/'
attrDir= r'/home/konata/Dataset/TED_MRI/T2/mask/origin_cut/'
sheetDir=r'/home/konata/Dataset/TED_MRI/T2.csv'

objDir=r'/home/konata/Dataset/TED_MRI/T2/seg_slice_3d_all_Jun_17/'

if not os.path.exists(objDir):
    os.mkdir(objDir)

files=os.listdir(baseDir)

filenames=pd.read_csv(sheetDir)

#x_yyyy-mm-dd_z.nii.gz，先按x排序，再按z排序
files = sorted(files, key=lambda x: (int(x.split('_')[0]), int(x.split('.')[0].split('_')[2])))

for index, row in filenames.iterrows():
    shape=row['shape_z']
    filename=row['id'].split('.')[0]
    #attr用于获得原3d图像的spacing，origin，direction属性
    attr=sitk.ReadImage(os.path.join(attrDir,row['id']))

    #读取第一张切片，获取切片的shape用来设定new_img的height和width
    img=sitk.ReadImage(os.path.join(baseDir,filename+'_0.nii.gz'))
    img=sitk.GetArrayFromImage(img)
    new_img=np.zeros((shape,img.shape[0],img.shape[1]))

    #根据sheet里的shape_z遍历切片按序号读取并加载入全零numpy数组new_img中
    for i in range(shape):
        img=sitk.ReadImage(os.path.join(baseDir,filename+'_'+str(i)+'.nii.gz'))
        img=sitk.GetArrayFromImage(img)
        new_img[i,:,:]=img
    
    #设定direction，spacing，origin属性并保存
    new_img=sitk.GetImageFromArray(new_img)
    new_img.SetDirection(attr.GetDirection())
    new_img.SetOrigin(attr.GetOrigin())
    new_img.SetSpacing(attr.GetSpacing())
    sitk.WriteImage(new_img,os.path.join(objDir,row['id']))
```
